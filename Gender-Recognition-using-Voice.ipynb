# Gender Voice Recognition 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("voice.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

## 1. Pre-processing of dataset

```python
df.dropna(subset=['label'], inplace=True)
```

```python
df.shape
```

```python
df = df[df['label'] != '']
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l = LabelEncoder()
```

```python
arr = l.fit_transform(df['label'])
```

```python
df['label'] = arr
```

```python
df.head()
```

```python
df.tail()
```

## 2. Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_data = df.drop(columns=['label'])
```

```python
y_data = df['label']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.4, random_state=42)
```

## 3. Building the model

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lor = LogisticRegression()
```

```python
lor.fit(x_train, y_train)
```

```python
pre = lor.predict(x_test)
```

```python
lor.score(x_test, y_test)
```

```python
from sklearn.metrics import f1_score
```

```python
f1_score(y_test, pre)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, pre)
```

## 4. Evaluating the model

```python
# !pip install matplotlib
```

```python
# !pip install seaborn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.metrics import confusion_matrix
```

### i. Using Logistic Regression

```python
cm_lor = confusion_matrix(y_test, pre)
```

```python
plt.figure(figsize = (12,8))
plt.title("Confusion matrix on test data using Logistic Regression")
sns.heatmap(cm_lor, annot=True, fmt='d', cmap=plt.cm.Reds, cbar=True, annot_kws={'size':14})
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

### ii. SVM-rbf

```python
from sklearn.svm import SVC
```

```python
clf = SVC(kernel = 'rbf')
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
svm_predictions = clf.predict(x_test)
```

```python
f1_score(y_test, svm_predictions)
```

```python
r2_score(y_test, svm_predictions)
```

```python
acc = accuracy_score(y_test, svm_predictions)
print(acc)
```

```python
cm_svm = confusion_matrix(y_test, svm_predictions)
```

```python
plt.figure(figsize = (8,8))
plt.title("Confusion matrix on test data using SVM-rbf")
sns.heatmap(cm_svm, annot=True, fmt='d', cmap=plt.cm.Blues, cbar=False, annot_kws={'size':14})
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```
